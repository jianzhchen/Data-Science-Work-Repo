## 1.generate the scatter plot of these two-dimensional data

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [10,20]
cov = [[3, 4], [4, 8]]  

samples = np.random.multivariate_normal(mean, cov, 5000).T
samples_x,samples_y=samples
plt.plot(samples_x, samples_y, 'x')
plt.axis('equal')
plt.show()


```

## 2.center the data

```python
samples_centered = (samples.T-[10,20]).T

```

## 3.generate the scatter plot of after centering the data

```python
samples_centered_x,samples_centered_y=samples_centered
plt.plot(samples_centered_x, samples_centered_y, 'x')
plt.axis('equal')
plt.show()
```

## 4.compute the sample covariance matrix and compare it with the acual covariance of the distribution

```python
cov_samples=np.cov(samples)
cov_samples_centered=np.cov(samples_centered)
print(cov_samples)
print(cov_samples_centered)
```

## 5.are these two covariance matrices identical? if not what is the best way to make them as close of as possible ?

The two covariance matrices are identical

## 6.compute the eigen vectors and values of the covariance matrix using np.linalg.eig; this way you decompose the covariance martix, i.e.$S = E DE^{-1}$ where $E$ is the eigen vector matrix and $D$ is a diagonal matrix whose elements are eigen values. If you have no knowledge of eigen vectors and values, don't worry. The main objective of the assignment is not about eigen vectors and values. We will talk about them later in the course.

```python
eigen_value,eigen_vector=np.linalg.eig(cov_samples_centered)
```

## 7.transform the data matrix by $E^T$, i.e. , $X_t = E^T X$

```python
samples_centered_t = eigen_vector.T.dot(samples_centered)
```

## 8.generate the scatter plot of the two-dimensional data $X_t$. What change you observe in the scatter plot and what is the implication of this transformation?

The data is centered vertically
The data is now axis aligned
The convarience matrix is now

$\left[
\begin{array}{cccc}
a & 0  \\
0 & b  \\
\end{array}
\right]$


```python
samples_centered_t_x,samples_centered_t_y=samples_centered_t
plt.plot(samples_centered_t_x, samples_centered_t_y, 'x')
plt.axis('equal')
plt.show()

print(np.cov(samples_centered_t))
```

## 9.transform the data matrix $X_t$ by $D^{-0.5}$, i.e. , $X_w = D^{-0.5} X_t$

```python
eigen_value_d=np.power(eigen_value,-0.5)
eigen_value_d=np.diag(eigen_value_d)

samples_centered_w=eigen_value_d.dot(samples_centered_t)
```

## 10.generate the scatter plot of the two-dimensional data $X_w$ .What change you observe in the scatter plot and what is implication of this transformation?

The data is centered around 0,0
The spread in y is equal to the spread in x
The convarience matrix is now

$\left[
\begin{array}{cccc}
1 & 0  \\
0 &1  \\
\end{array}
\right]$


```python
samples_centered_w_x,samples_centered_w_y=samples_centered_w
plt.plot(samples_centered_w_x, samples_centered_w_y, 'x')
plt.axis('equal')
plt.show()

print(np.cov(samples_centered_w))
```
